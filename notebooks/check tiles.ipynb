```python
import pandas as pd
import os
import cv2
import numpy as np

from tqdm import tqdm_notebook
```

```python
root_folder = "/home/user/projects/agrivision/data/Agriculture-Vision"
```

```python
dataframe = pd.DataFrame(columns=['name', 'ds_part'])

for ds_part in ['train', 'val', 'test']:
    ds_part_path = os.path.join(root_folder, ds_part, 'images', 'rgb')
#     ds_part_path = os.path.join(root_folder, ds_part)
    dataset = os.listdir(ds_part_path)
    df = pd.DataFrame(columns=['name', 'ds_part'])
    df['name'] = dataset
    df['ds_part'] = ds_part
    dataframe = dataframe.append(df)
```

```python
train_val_df = dataframe[dataframe['ds_part'].isin(['train', 'val'])]
```

```python
labels = ['double_plant',
 'planter_skip',
 'weed_cluster',
 'waterway',
 'cloud_shadow',
 'standing_water']
for num, row in tqdm_notebook(train_val_df.iterrows(), total=len(train_val_df)):
    for label in labels:
        img_path = os.path.join(root_folder, row['ds_part'], 'labels', label, row['name'].replace(".jpg", ".png"))
        # print(img_path)
#         try:
        img = cv2.imread(img_path)
        # img.shape
#         if not isinstance(img, np.ndarray):
#             print(img_path)
#             break
        num_pixels = np.sum(img)
        # print(num_pixels)
        train_val_df.loc[num, label] = num_pixels
#         except Exception as e:
#             print(e)
#             print(img_path)
```
